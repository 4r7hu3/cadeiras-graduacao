A validação cruzada (*cross-validation*), de um modo geral, é um conjunto de técnicas utilizadas para medir e quantificar a performance de estimadores e/ou modelos de machine learning no contexto da Ciência de Dados.

Tais técnicas consistem, na verdade, de métodos de reamostragem, pois são baseadas na divisão de um conjunto de dados em vários subconjuntos, que serão utilizados para treinar e testar o modelo separadamente, em várias iterações.

Com a validação cruzada, é possível **identificar** problemas como *underfitting* e *overfitting*, além do viés de seleção, a fim de saber como tal modelo irá se comportar com novas observações.

### Problemas de modelagem

#### *Underfitting*

Do inglês, *underfitting* refere-se a algo que se ajusta de maneira insuficiente. Na Ciência de Dados, este termo se traduz como a baixa captura de padrões por um modelo, ou seja, o mesmo não consegue captar bem como os dados se comportam, de modo que também possui baixa performance quando usado em novas observações.

#### *Overfitting*

Aqui, temos a situação inversa. Ocorre que o modelo se ajusta tão bem aos dados, que captura, inclusive, seus ruídos e pontos de influência. Assim, para modelos com essa característica, há um ótimo ajuste aos dados de treino, mas uma baixa performance para dados de teste (novas observações).

#### Viés de seleção
Imagine que queremos fazer um estudo sobre o tempo médio gasto por pessoas assistindo televisão todos os dias. Se uma amostra para tal pesquisa fosse retirada de uma casa de repouso, teríamos viés nos nossos dados, pois a amostra não é representativa (pessoas em casas de repouso tendem a passar mais tempo vendo TV). Assim, um modelo com viés de seleção/amostragem não é adequado, pois pode distorcer os resultados e levar a conclusões errôneas.

### Tipos de validação


#### *K-fold*
O método k-fold divide o conjunto de dados em k partes iguais e usa uma parte como conjunto de validação (teste) e as outras k-1 como conjunto de treinamento. O processo é repetido k vezes até que cada parte seja testada uma vez.

traintest.png

* Vantagens:
  + Uso eficiente dos dados para treinamento
  + Previne overfitting
  + Reduz a variância na estimativa de desempenho

* Desvantagens:
  + Aumento do custo e tempo computacional
  + Limitado em certos tipos de dados
  + Viés em dados desequilibrados

#### *Holdout*

Este método consiste na divisão aleatória de um conjunto de dados em um conjunto de treinamento e um conjunto de validação. Ao **contrário** do método *k-fold*, esse método é dividido em duas partes, uma com aproximadamente 70% do conjunto de dados para o treinamento e os 30% restantes para validação.

holdout.png

* Vantagens:
  + Fácil de implementar
  + Eficiência e Rapidez

* Desvantagens:
  + Sensível à escolha da divisão
  + Uso ineficiente de dados disponíveis


### *Monte Carlo*

Quando se utiliza um modelo estatístico para descrever um sistema, é possível que o modelo seja
complexo demais, ou então, não permita uma solução analítica. Nesse caso, a simulação computacional pode
ser considerada uma ferramenta de grande valia na obtenção de uma resposta para um problema particular.
Quando o modelo envolve amostragem aleatória de uma distribuição probabilística, o método é designado
Simulação de Monte Carlo.

mc.png

Se assemelha a uma combinação dos métodos *holdout* e *k-fold*. No método *holdout*, ocorre uma única divisão dos dados em conjuntos de treinamento e teste, enquanto o *k-fold* divide os dados em k subconjuntos (folds) para treinamento e validação, por outro lado a validação de Monte Carlo amplia esses conceitos.

Na validação de Monte Carlo, o conjunto de dados é dividido aleatoriamente em conjuntos de treino e teste várias vezes, criando múltiplas divisões diferentes. Cada divisão é considerada uma realização ou iteração do experimento.

montec.png

* Vantagens:
  + Estimativa mais robusta
  + Maior confiabilidade
  + Redução do viés

* Desvantagens:
  + Custo computacional elevado
  + Maior tempo de execução
  + Complexidade de implementação

### *Leave-one-out*

Para cada iteração, utiliza-se um único ponto de dados como conjunto de teste, enquanto todos os outros pontos de dados são usados como conjunto de treinamento. Esse processo é repetido para cada ponto de dados no conjunto de dados, o que significa que a quantidade de iterações é igual ao número total de pontos de dados.

loocv.png

* Vantagens:
  + Utilização máxima dos dados
  + Menor viés na estimativa

* Desvantagens:
  + Custo computacional elevado
  + Sensível a *outliers*
  + Viés em dados desequilibrados

### **Métricas de validação**

Métricas de validação são medidas quantitativas utilizadas para verificar a performance do modelo/estimador em questão de forma rápida e concisa. Elas podem revelar a efetividade de um modelo, e são bastante úteis na comparação entre modelos, pois são capazes de nortear os pesquisadores e suas decisões sobre o uso de cada um. Alguns exemplos de métricas, aplicadas tanto na Estatística quanto na Ciência de Dados, são:

 - $R^2\, \text{e}\, R^2_{a.}$: Coeficiente de Determinação e Coeficiente de Determinação Ajustado, medem a proporção da variabilidade de uma variável resposta que é explicada por um modelo de regressão;
 - $MAE$: O erro absoluto médio mostra a falta de acurácia esperada (média) das previsões;
 - $MAPE$: O erro percentual médio absoluto mede a proporção média da falta de acurácia esperada das previsões, e é uma alternativa ao MAE;
 - $MSE$ : Erro Quadrático Médio, é a forma mais poderosa de medir a qualidade de estimadores através da variância e viés dos mesmos;
 - $RMSE$: A raíz quadrada do MSE (alternativa a essa medida) revela a "distorção" na mesma unidade da variável;
 - $AIC$: Critério de Informação de Akaike, baseado na teoria da informação. É utilizado para comparar e selecionar modelos, sendo que valores mais baixos de AIC indicam maior qualidade e simplicidade. Utiliza-se do critério do parcimônia.
 - $BIC$: Critério de informação bayesiano, equilibra a qualidade do ajuste do modelo com a penalização pela complexidade do modelo, favorecendo modelos mais simples, o objetivo é encontrar o equilíbrio ideal entre precisão e simplicidade ao escolher entre diferentes modelos estatísticos. Valores mais baixos de BIC indicam uma melhor qualidade do modelo.

# Exemplos

## **Métricas em Leave-one-out**

```r
install.packages("caret")
library(caret)

#-=-=- Leave-one-out
set.seed(19)
x = rnorm(50, 5, 2)
y = 2*x + rnorm(50)
dados = data.frame(x,y)

modelo_loocv = train(y ~ x, method = "lm", data = dados, trControl = trainControl(method = "LOOCV"))

loocv = c(modelo_loocv$results[-1]) #valores do train

#-=-=- Modelo linear
m = lm(y ~ x)

y_pred = predict(m)
MAE = mean(abs(y - y_pred))
RMSE = sqrt(mean((y - y_pred)^2))
R2 = summary(m)$r.squared
modelo_lm = c(RMSE,R2,MAE)

#-=-=- Resultados
cbind(loocv, modelo_lm)
```

## **Simulação Monte Carlo**

Neste exemplo, estamos utilizando o modelo de simulação de Monte Carlo para estimar as médias dos parâmetros $\hat{\beta_0}, \hat{\beta_1}$ e $\hat{\beta_2}$, viés de estimativa, variância e erro quadrático médio (MSE) da seguinte função de regressão linear:  $$\boldsymbol{\hat{Y}} = 5+1,2 \times X_1 + 3 \times X_2$$

```r
set.seed(521353)
n = c(200,1000,2000,5000,10000) # tamanhos de amostras a serem simulados
NREP = 100 # numero de iteracoes
```

```r
betaMatrix = matrix(NA,100,3)
listMSE = c()
```

```r
for(i in 1:length(n)){

  x1 = rnorm(n=n[i],mean = 2, sd = 0.4) # amostra normal(2, 0.16)
  x2 = rnorm(n=n[i],mean = 1,sd = 0.1) # amostra normal(1, 0.01)

  for(j in 1:NREP){

    error = rnorm(n = n[i], mean = 0,sd = 1) # sabemos em MRL E~N(0, 1)
    y =  5 + 1.2*x1 + 3*x2 + error # modelo de regressao que esta sendo analisado
    lm.out = lm(y~x1+x2) # criacao do modelo
    betaMatrix[j,] = lm.out$coefficients
    mysummary = summary(lm.out)
    listMSE[j] = mysummary$sigma^2 # pegando cada MSE de cada iteração
  }

  beta_mean = apply(betaMatrix,2,mean)
  estimation_bias = beta_mean - c(5,1.2,3) # estimando o viés
  beta_var = apply(betaMatrix,2,var) # variancia dos estimadores
  beta_mse = estimation_bias^2 + beta_var # Erro quadrático médio dos estimadores
  cat("Tamanho da Amostra: ",n[i],"\n")
  cat("Media_Beta = ",beta_mean,"\nVies_Beta = ",estimation_bias,"\nVarBeta = ",beta_var,"\nMSE_Beta = ",beta_mse,"\nModel MSE = ",mean(listMSE))
  cat("\n\n")
}
```

## **Conclusões**

*   As estimativas médias dos parâmetros são muito próximas em valor ao verdadeiro valor dos parâmetros.
*   A variância de cada uma das estimativas dos parâmetros diminui com o aumento do tamanho da amostra, enquanto o viés não apresenta uma relação clara com o tamanho da amostra.

* O EQM está diminuindo com o aumento do tamanho da amostra. Isso ocorre porque o impacto da diminuição da variância com o aumento do tamanho da amostra supera o impacto da variação aleatória devido à presença do termo de viés ao quadrado $EQM[\hat{\theta}] = Var[\hat{\theta}] + B^2[\hat{\theta}]$.

* A Média do *MSE* do Modelo Estimado é próxima a 1 em cada simulação.

# Retorno


## Ação

Dada uma lista com os preços dos fechamentos de uma determinada ação, calcule os retornos entre os fechamentos. Coloque os retornos em uma nova lista.

$$ retorno =  \frac{fechamento\_atual}{fechamento\_anterior} - 1 $$

Crie uma função `retorno(precos)` que recebe uma lista de preços (`precos`) e retorna uma lista de retornos.

## Observação
Não usar funções prontas. Fazer usando laços/repetição.


```python
# Crie a função aqui.
def retorno(precos):
  retornos = []

  for i in range(1, len(precos)):
    retorno = (precos[i] / precos[i-1]) - 1
    retornos.append(retorno)

  return retornos
```

```python
## Casos de teste. Não modificar.

print(retorno([37.8, 36.8, 38.0, 39.5]))
print(retorno([100.20, 98.80, 99.50, 102.80, 107.20]))


# Saída esperada
# [-0.02645502645502651, 0.03260869565217406, 0.03947368421052633]
# [-0.013972055888223589, 0.007085020242914908, 0.03316582914572863, 0.04280155642023353]

```

# L1 - Filtro

## Ação

Crie uma função filtro(l) que recebe uma lista de inteiros l. Sua função deve retornar uma cópia da lista de entrada, mas sem os números que são múltiplos de 3.

## Observação
Não usar funções prontas. Fazer usando laços/repetição.


```python
# Crie a função aqui.
def filtro(l):
  l_copia = l.copy()

  for i in l:
    if i % 3 == 0:
      l_copia.remove(i)

  return l_copia
```

```python
## Casos de teste. Não modificar.

print(filtro([2,3,6]))
print(filtro([0,0,1,0]))
print(filtro([2,4,2,6,2,0,2,2]))

# Saída esperada
# [2]
# [1]
# [2, 4, 2, 2, 2, 2]
```

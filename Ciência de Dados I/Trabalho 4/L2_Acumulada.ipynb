# L2 -Função Acumulada

## Motivação

Imagine que você é o gerente financeiro de uma pequena empresa de tecnologia chamada TechNova. A empresa está crescendo rapidamente e recentemente lançou um novo produto inovador no mercado. No entanto, com o crescimento acelerado, surgem desafios financeiros.

Seu CEO, Ana, está preocupada com a gestão do fluxo de caixa, pois a empresa precisa garantir que terá dinheiro suficiente para pagar fornecedores, funcionários e investir em novas oportunidades. Ana pede sua ajuda para analisar o fluxo de caixa acumulado dos últimos meses.

Por exemplo, temos o seguinte fluxo de caixa:
- Janeiro: +50
- Fevereiro: -20
- Março: +30
- Abril: -10
- Maio: +40
- Junho: -15

Ela quer que você gere o fluxo acumulado mês a mês. Em cada mês, você calcula o somatório dos fluxos do primeiro mês até o atual:
- Janeiro: +50
- Fevereiro: +30
- Março: +60
- Abril: +50
- Maio: +90
- Junho: +75

## Ação

Crie uma função acumulada(l) que recebe uma lista de float l representando os fluxos de caixa mensais da empresa. Sua função deve retornar uma nova lista com os valores acumulados mês a mês.


## Observação
Não usar funções prontas. Fazer usando laços/repetição.


```python
# Crie a função aqui.
def acumulada(l):
  ac = l

  for i in range(1, len(l)):
    ac[i] = ac[i] + ac[i-1]

  return ac
```

```python
## Casos de teste. Não modificar.

print(acumulada([1, 2, 3, 4, 5]))
print(acumulada([10, -20, -30, 40, 50]))
print(acumulada([-10, -20, -40, 60]))

# Saída esperada:
# [1, 3, 6, 10, 15]
# [10, -10, -40, 0, 50]
# [-10, -30, -70, -10]
```

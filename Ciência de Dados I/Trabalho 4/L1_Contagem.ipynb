# L1 - Contagem



## Motivação

Na saga dos Vingadores, o Ultron é um inimigo capaz de fazer cópias de sí mesmo. João é um grande fã da Saga e possui um enorme poster que contém diversas copias do Ultron junto aos heróis. João gostaria de saber quantos Ultrons estão presentes no poster.


## Ação

Crie uma função contagem(l, x) que recebe uma lista de inteiros l e um inteiro x. Sua função deve retornar o número de vezes que o inteiro x se repete na lista.


## Observação
Não usar funções prontas. Fazer usando laços/repetição.


```python
# Crie a função aqui.
def contagem(l, x):
  cont = 0

  for i in l:
    if i == x:
      cont += 1
  return cont
```

```python
## Casos de teste. Não modificar.

print(contagem([2,3,6], 5))
print(contagem([0,0,1,0], 0))
print(contagem([2,4,2,6,2,0,2,2], 2))

# Saída esperada
# 0
# 3
# 5

```

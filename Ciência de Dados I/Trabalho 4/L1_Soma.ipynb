# L1 - Soma


## Ação

Crie uma função soma(l) que recebe uma lista de inteiros. Sua função deve retornar a soma de todos os elementos de l.


## Observação
Não usar funções prontas. Fazer usando laços/repetição.


```python
# Crie a função aqui.
def soma(l):
  soma = 0

  for i in l:
    soma += i

  return soma
```

```python
## Casos de teste. Não modificar.

print(soma([4,2,6]))
print(soma([8,1,2,9]))
print(soma([2,2,6,2,2]))

# Saída esperada
# 12
# 20
# 14
```

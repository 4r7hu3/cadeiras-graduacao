# Detecção de Arbitragem entre Mercados

Você está trabalhando como analista financeiro em uma corretora e precisa identificar oportunidades de arbitragem entre dois mercados financeiros que negociam o mesmo ativo. A arbitragem ocorre quando um ativo é vendido a um preço mais baixo em um mercado e vendido a um preço mais alto em outro, gerando um lucro sem risco.

Sua tarefa é criar um programa que detecte oportunidades de arbitragem ao comparar os preços de um ativo em dois mercados diferentes.

## Tarefa

Crie uma função chamada `detectar_arbitragem(mercado_a, mercado_b)` que recebe duas listas de preços diários de um ativo em dois mercados diferentes e retorna uma lista de string indicando a operação a ser feita usando a seguinte lógica:

- Se o preço do ativo no Mercado A for menor que o preço no Mercado B, há uma oportunidade de comprar no Mercado A e vender no Mercado B. Nesse caso deve-se indicar a string "comprar no mercado A e vender no mercado B".
- Se o preço do ativo no Mercado B for menor que o preço no Mercado A, há uma oportunidade de comprar no Mercado B e vender no Mercado A. Nesse caso deve-se indicar a string "comprar no mercado B e vender no mercado A".
- Se o preço do ativo for igual nos dois mercados, não há oportunidade de arbitragem. Nesse caso deve-se indicar a string "sem oportunidade de arbitragem".

## Exemplo

Entrada:  
`mercado_a = [100.0, 102.5, 98.0, 105.0, 100.0]`  
`mercado_b = [101.5, 101.0, 99.5, 104.0, 100.0]`  

```
resultado = ['comprar no mercado A e vender no mercado B',
            'comprar no mercado B e vender no mercado A',
            'comprar no mercado A e vender no mercado B',
            'comprar no mercado B e vender no mercado A',
            'sem oportunidade de arbitragem']
```


## Observação
Não usar funções prontas, faça uso de laços/repetição e listas.


````python
# Crie a função aqui.
def detectar_arbitragem(mercado_a, mercado_b):

  recomendacao = []

  for i in range(len(mercado_a)):
    if mercado_a[i] < mercado_b[i]:
      recomendacao.append('comprar no mercado A e vender no mercado B')
    elif mercado_b[i] < mercado_a[i]:
      recomendacao.append('comprar no mercado B e vender no mercado A')
    else:
      recomendacao.append('sem oportunidade de arbitragem')

  return recomendacao
````

````python
# Exemplo de uso
print(detectar_arbitragem([100.0, 102.5, 98.0, 105.0, 100.0], [101.5, 101.0, 99.5, 104.0, 100.0]))

# Saída esperada:
# ['comprar no mercado A e vender no mercado B',
# 'comprar no mercado B e vender no mercado A',
# 'comprar no mercado A e vender no mercado B',
# 'comprar no mercado B e vender no mercado A',
# 'sem oportunidade de arbitragem']
````

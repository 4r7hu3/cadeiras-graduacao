# L1 - Mediana


## Motivação

A mediana é uma medida de tendência central que indica o valor central de um conjunto de dados ordenados. Em outras palavras, é o número que separa a metade superior da metade inferior dos dados.

Para calcular a mediana, siga estas etapas:

- Ordenar os dados: Coloque todos os números em ordem crescente.
- Se número ímpar de elementos: Se o conjunto de dados tiver um número ímpar de elementos, a mediana é o número do meio.
- Se número par de elementos: Se o conjunto de dados tiver um número par de elementos, a mediana é a média dos dois números do meio.

Por exemplo, a mediana da lista [5, 3, 2, 4, 1] é 3. A mediana da lista [5, 3, 2, 6, 1, 4] é 3.5.


## Ação

Crie uma função mediana(l, x) que recebe uma lista de inteiros l e retorna a mediana da lista.

## Observação
Não usar funções prontas. Fazer usando laços/repetição.


```python
# Crie a função aqui.
def mediana(l):

  for i in range(len(l)):
    for j in range(0,len(l)-1):
      if l[j] > l[j+1]:
        l[j], l[j+1] = l[j+1], l[j]

  if len(l) % 2 != 0:
    return l[int(len(l)/2)]
  else:
    return (l[int(len(l)/2)] + l[int(len(l)/2)-1])/2
```

```python
## Casos de teste. Não modificar.

print(mediana([5, 3, 2, 4, 1])) # 3
print(mediana([5, 3, 2, 6, 1, 4])) # 3.5
print(mediana([48, 23, 10, 39, 49])) # 39
print(mediana([48, 23, 10, 39, 49, 83])) # 44

```

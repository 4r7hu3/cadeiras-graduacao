# L2 - Média Acumulada

## Motivação


O novo desafio é o seguinte. Sua chefe quer entender como o fluxo de caixa tem se comportado em média. Assim, dado o fluxo de caixa mensal da sua empresa, sua chefe quer que você calcule a média acumulada. A cada mês do primeiro mês, você deve calcular a média do primeiro mês até o mês atual.

Por exemplo, temos o seguinte fluxo de caixa:
- Janeiro: +50
- Fevereiro: -20
- Março: +30
- Abril: -10
- Maio: +40
- Junho: -15

Assim, a lista de médias acumuladas é o seguinte
- Janeiro: +50
- Fevereiro: +15
- Março: +20
- Abril: +12.5
- Maio: +18
- Junho: +12.5

## Ação

Crie uma função media_acumulada(l) que recebe uma lista de float l representando os fluxos de caixa mensais da empresa. Sua função deve retornar uma nova lista com as médias acumuladas mês a mês.


## Observação
Não usar funções prontas. Fazer usando laços/repetição.


```python
# Crie a função aqui.
def media_acumulada(l):
  m_acum = [l[0]]
  cont_mes = 2

  for i in range(1, len(l)):
    l[i] += l[i-1]
    m_acum.append(l[i]/cont_mes)
    cont_mes += 1

  return m_acum
```

```python
## Casos de teste. Não modificar.

print(media_acumulada([1, 2, 3, 4, 5]))
print(media_acumulada([10, -20, -30, 40, 50]))
print(media_acumulada([-10, -20, -40, 60]))

# Saída esperada:
# [1, 1.5, 2, 2.5, 3]
# [10, -5, -13.33, 0, 10]
# [-10, -15, -23.33, -2.5]
```

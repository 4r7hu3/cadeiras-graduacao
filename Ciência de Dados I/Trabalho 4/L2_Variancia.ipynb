# L2 - Variância

## Motivação

A variância é uma medida estatística que indica o quanto os preços de uma ação variam em relação à média desses preços. Em termos simples, ela mostra a dispersão dos preços ao longo do tempo.

Uma alta variância significa que os preços da ação têm grandes flutuações, indicando maior volatilidade e, possivelmente, maior risco. Por outro lado, uma baixa variância sugere que os preços são mais estáveis e menos sujeitos a grandes mudanças.

Fórmula da variância amostral:

$$ s^2 = \sum_{i=1}^n \frac{(x_i -\bar{x})^2}{n-1} $$

onde $\bar{x}$ é a média de x

## Ação

Crie uma função variancia(l) que recebe uma lista de float representando preços de fechamento de um dado ativo financeiro. Sua função deve retornar a variância dos preços de l. Além disso, crie também a função media(l) que recebe uma lista e retorna a média da lista. A função variancia deve chamar a função media.


## Observação
Não usar funções prontas. Fazer usando laços/repetição.


```python
# Crie a função aqui.
def variancia(l):
  n = len(l)
  soma = 0

  for i in l:
    soma+= i

  media = soma/n
  soma = 0

  for i in range(n):
    soma+= (l[i]-media)**2

  return soma/(n-1)
```

```python
## Casos de teste. Não modificar.

TSLA34 = [38.03, 38.12, 39.01, 35.79, 36.13, 36.53, 36.48, 36.86, 36.82, 36.87]
print(variancia(TSLA34))

AAPL34 = [61.50, 61.15, 61.11, 61.47, 61.50, 61.60, 61.50, 62.55, 61.60, 60.55]
print(variancia(AAPL34))

GOOGL34 = [78.60, 78.50, 79.00, 79.80, 79.90, 76.90, 78.24, 78.50]
print(variancia(GOOGL34))
```

# Preços - Retorno


## Descrição

Dada uma lista de números reais $r_1, r_2, \ldots, r_n$ representando os retornos diários de uma ação, e um número real $p_0$ representando o preço inicial da ação, queremos calcular o preço final da ação após $n$ dias. O preço final da ação é dado por $p_n = p_0 \times (1 + r_1) \times (1 + r_2) \times \ldots \times (1 + r_n)$.

## Ação

Crie uma função chamada `preco_final(retornos, preco_inicial)` que recebe uma lista de números reais `retornos` e um número real `preco_inicial` e retorna o preço final da ação após `n` dias.

```python
# Crie a função aqui.
def preco_final(retornos, preco_inicial):
    preco_final = preco_inicial

    for retorno in retornos:
        preco_final *= (1 + retorno)

    return preco_final
```

```python
## Casos de teste. Não modificar.

print(preco_final([0.03, 0.07, -0.1, 0.2, -0.5], 100))
print(preco_final([0.05, 0.1, 0.5, -0.5, 0.3], 500))
print(preco_final([0.07, 0.02, 0.3, -0.2, 0.5], 1000))



```

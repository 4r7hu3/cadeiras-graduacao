# Máximo

## Motivação

Dada uma lista com os preços dos fechamentos de uma determinada ação, calcule o valor máximo que a ação atingiu.

## Ação

Crie uma função chamada `maximo(lista)` que recebe uma lista de float representando os preços de fechamento e retorna o maior valor encontrado na lista.

## Observação
Não usar fórmulas prontas como max() ou amax(). Fazer usando laços/repetição.


```python
# Crie a função aqui.
def maximo(lista):
  max = lista[0]

  for i in lista:
    if i > max:
      max = i

  return max
```

```python
## Casos de teste. Não modificar.

print(maximo([1, 2, 6, 2, 0]))
print(maximo([7.3, 7.2, 6.1, 2.2, 0.0]))
print(maximo([-1.0, -2.0, -6.0, -2.0, 0.0]))

# Saídas esperadas
# 6
# 7.3
# 0.0
```

## Tarefa

Crie a função nota(ap1, ap2, final) que recebe 3 números float e retorna True caso o aluno tenha sido aprovado e False caso contrário.

Obs: descarte a nota da final se ela não for necessária.


````python
# Crie a função aqui
def nota(ap1, ap2, final):
    media = (ap1 + ap2)/2

    if media >= 7:
        return True
    elif media < 4:
        return False
    elif media >= 4 and media < 7:
        mediaFinal = (media + final)/2
        if mediaFinal >= 5:
            return True
        else:
            return False
    else:
        False
````

````python
# Código de teste. Não modifique!
print(nota(8, 8, 9))
print(nota(8, 2, 9))
print(nota(4, 2, 9))
print(nota(4, 8, 3))

# Saída esperada:
# True
# True
# False
# False
````

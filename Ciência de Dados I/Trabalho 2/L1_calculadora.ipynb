# Calculadora Numérica


## Motivação

Quem nunca numa prova de matemática ou física no ensino médio escutou um coleguinha dizer: "Professor, pode usar a calculadora?" que atire a primeira pedra.

### Ação

Implemente uma função calculadora(n1, n2, operacao) que recebe dois inteiros e um caractere ('+', '-', '\*' ou '/'). Essa função retorna o resultado da operação.


```python
# Crie a função aqui
def calculadora(n1, n2, operacao):

    if operacao == '+':
        return n1 + n2
    elif operacao == '-':
        return n1 - n2
    elif operacao == '*':
        return  n1 * n2
    else:
        return n1 / n2
```

```python
# Código de teste. Não modificar.
print(calculadora(5, 2, '/'))
print(calculadora(3, 5, '+'))
print(calculadora(7, 9, '/'))

# Saída esperada:
# 2.5
# 8
# 0.7777777
```

## Ação

Crie a função maior(a, b, c, d) que recebe 4 números e retorna o maior deles

```python
# Crie sua função aqui
def maior(a, b, c, d):

   if a >= b and a>= c and a>= d:
       return a
   elif b >= a and b >= c and b>= d:
       return b
   elif c >= b and c >= a and c >= d:
       return c
   else:
       return d
```

```python
print(maior(3,3,2,2))
print(maior(5,1,8,3))

# Saída esperada
# 3
# 8
```

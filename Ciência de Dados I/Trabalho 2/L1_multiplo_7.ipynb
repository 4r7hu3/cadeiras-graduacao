## Ação

Implemente uma função multiplo_7(n) que recebe um número inteiro e retorna True caso ele seja múltiplo de 7 e False caso contrário.


```python
# Cria a função aqui
def multiplo_7(n):
    
    if n % 7 == 0:
        return True
    else:
        return False
```

```python
# Código de teste. Não modificar.

print(multiplo_7(3))
print(multiplo_7(14))
print(multiplo_7(21))
print(multiplo_7(0))
print(multiplo_7(45))
print(multiplo_7(-20))
print(multiplo_7(-21))

# Saída esperada:
# False
# True
# True
# True
# False
# False
# True
```

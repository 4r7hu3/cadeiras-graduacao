## Motivação

Três competidores olímpicos de arremesso de disco tinham desempenhos muito parecidos. Constantemente eles jogavam os discos à mesma distância. Decidiram então criar um algoritmo que informava quantos deles tinham conseguido o mesmo resultado.

## Ação

Crie a função iguais(a,b,c) que recebe 3 inteiros e retorna a quantidade de valores iguais dentre os 3.


```python
# Cria sua função aqui
def iguais(a,b,c):
    
    if a == b and b == c:
        return 3
    elif a != b and b != c and a != c:
        return 0
    else:
        return 2
```

```python
# Código de teste. Não modifique!

print(iguais(1, 4, 4))
print(iguais(1, 5, 4))
print(iguais(1, 1, 1))
print(iguais(1, 1, 2))
print(iguais(1, 2, 1))

# Saída esperada:
# 2
# 0
# 3
# 2
# 2

```

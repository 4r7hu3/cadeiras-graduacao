# Cálculo da Alavancagem Financeira de uma Empresa

Você foi contratado como analista financeiro para avaliar o nível de alavancagem financeira de uma empresa. A alavancagem financeira é uma medida de quanto a empresa está utilizando capital de terceiros (dívidas) em relação ao seu capital próprio para financiar suas operações. Quanto maior a alavancagem, maior é o risco financeiro da empresa, pois ela depende mais de dívida para gerar lucro.

A fórmula para calcular a alavancagem financeira de uma empresa é:

$$ Alavancagem \; Financeira = \frac{Dívida \; Total}{Capital \; Próprio} $$

Onde:
- Dívida Total: soma das dívidas da empresa (empréstimos, financiamentos, debêntures, etc.).
- Capital Próprio: patrimônio líquido da empresa (capital dos acionistas).

## Tarefa

Crie uma função chamada `alavancagem_financeira(divida_total, capital_proprio)` que recebe dois parâmetros: `divida_total` e `capital_proprio`. A função deve retornar uma string indicando o nível de alavancagem financeira da empresa com base na lógica a seguir:

- Baixa, se alavancagem financeira for menor ou igual a 1.
- Moderada, se alavancagem financeira for maior que 1 e menor ou igual a 2.
- Alta, se alavancagem financeira for maior que 2.


```python
def alavancagem_financeira(divida_total, capital_propio):
    alavancagem = divida_total/capital_propio

    if alavancagem <= 1:
        return 'Baixa'
    elif alavancagem > 1 and alavancagem <= 2:
        return 'Moderada'
    else:
        return 'Alta'
```

```python
# Caso de teste. Não modificar

print(alavancagem_financeira(2000, 3000))
print(alavancagem_financeira(4000, 3000))
print(alavancagem_financeira(7000, 3000))

# Resultado esperado:
# Baixa
# Moderada
# Alta

```

## Motivação

A mãe de João entrou em seu quarto e ficou furiosa por causa da bagunça. Ela deu um grito bem alto e acordou o coitado do João que estava no terceiro sono. Ele sonhando que estava contando as ovelhas que conseguiam pular a cerca da fazenda de seu pai, mas no sonho, ele só conseguia contar de 60 em 60.

Após cair da cama, sua mãe lhe disse que se ele não se levantasse em um segundo e arrumasse seu quarto inteiro ficaria de castigo até o próximo ano bissexto, primo e quadrado perfeito.

## Ação

Crie a função proximo(hora, minuto, segundo) que recebe 3 inteiros e retorna três inteiro indicando a hora, minuto e segundo adicionado de um segundo.

```python
# Crie sua função aqui
def proximo(hora, minuto, segundo):
    segundo_adicionado = segundo + 1

    if segundo_adicionado >= 60:
        segundo_adicionado = 0
        minuto_adicionado = minuto + 1
        if minuto_adicionado >= 60:
            minuto_adicionado = 0
            hora_adicionada = hora + 1
            if hora_adicionada == 24:
                return 0, minuto_adicionado, segundo_adicionado
            else:
                return hora_adicionada, minuto_adicionado, segundo_adicionado
        else:
            return hora, minuto_adicionado, segundo_adicionado
    else:
        return hora, minuto, segundo_adicionado
            
```

```python
# Código de teste. Não modificar.

print(proximo(12, 10, 11))
print(proximo(12, 10, 59))
print(proximo(12, 59, 59))
print(proximo(23, 59, 59))

# Saída esperada:
# (12, 10, 12)
# (12, 11, 00)
# (13, 00, 00)
# (00, 00, 00)
```

No zerim ou um americano cada participante coloca quantos dedos quiser na roda. Um dos participantes soma os dedos e aponta para cada participante enquanto conta. Ganha aquele cujo dedo apontar no último número.

Suponha que existem 4 jogadores. O contador, o jogador 1, aponta pra cima e diz "zero", e então aponta pra si mesmo e diz "um", aponta pro jogador 2 e diz "dois", pro jogador 3 e diz "três", pro jogador 4 e diz "quatro", volta a apontar pra si e diz "cinco" e assim até que os números acabem.

## Tarefa

Crie um função zerinho_ou_um(jog1, jog2, jog3, jog4) que recebe quatro inteiros representando as jogadas de cada jogador. Essa função deve retorna a string "nenhum" se nenhum jogador ganhou ou "jog1", "jog2", "jog3" ou "jog4" indicando qual jogador ganhou.

```python
# Crie sua função aqui
def zerinho_ou_um(jog1, jog2, jog3, jog4):

    if (jog1+jog2+jog3+jog4) == 0:
        return 'nenhum'
    elif (jog1+jog2+jog3+jog4)%4 == 0:
        return 'jog4'
    elif (jog1+jog2+jog3++jog4)%4 == 0.25 or (jog1+jog2+jog3++jog4) == 1:
        return 'jog1'
    elif (jog1+jog2+jog3++jog4)%4 == 0.5 or (jog1+jog2+jog3++jog4) == 2:
        return 'jog2'
    else:
        return 'jog3'
```

```python
# Código de teste. Não moficar.

print(zerinho_ou_um(0, 0, 0, 0))
print(zerinho_ou_um(0, 0, 2, 0))
print(zerinho_ou_um(2, 2, 2, 2))

# Saída esperada
# nenhum
# jog2
# jog4
```

# 3 ou 5

## Ação

Implemente uma função tres_ou_cinco(n) que recebe um número inteiro. Essa função retorna True caso o número seja 3 ou 5 e retorna False caso contrário.

Use somente um if dentro da função.

```python
# Crie a função aqui
def tres_ou_cinco(n):
    if n == 3 or n == 5:
        return True
    else:
        return False
```

```python
# Código de teste. Não modificar.
print(tres_ou_cinco(3))
print(tres_ou_cinco(12))
print(tres_ou_cinco(-3))
print(tres_ou_cinco(-7))
print(tres_ou_cinco(5))

# Saída esperada:
# True
# False
# False
# False
# True
```

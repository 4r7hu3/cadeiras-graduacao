## Motivação

Já colocou o despertador no seu celular e ele lhe informou que você só tinha algumas poucas horas de sono?
Quem mandou dormir às 2 da manhã e ter que acordar às 5:30!

## Ação

Crie a função tempo_despertador(hora_atual, minuto_atual, segundo_atual, hora_despertador, minuto_despertador, segundo_despertador) que recebe a hora, minuto e segundos tanto da hora atual como da hora do despertador e retorne quantas horas minutos e segundos você tem de sono.

Se o segundo horário for menor que o primeiro, compreenda que é no outro dia. Exemplo, se você dormir 23:00 e acorda às 6:00 então você tem 7:00 horas de sono.

```python
# Crie sua função aqui
def tempo_despertador(hora_atual, minuto_atual, segundo_atual, hora_despertador, minuto_despertador, segundo_despertador):   

    if segundo_despertador < segundo_atual:
        segundo_despertador = segundo_despertador + 60
        minuto_despertador = minuto_despertador - 1 

    s = segundo_despertador - segundo_atual

    if minuto_despertador < minuto_atual:
        minuto_despertador = minuto_despertador + 60
        hora_despertador = hora_despertador - 1

    m = minuto_despertador - minuto_atual

    if hora_despertador < hora_atual:
        hora_despertador = hora_despertador + 24

    h = hora_despertador - hora_atual

    return h, m, s
```

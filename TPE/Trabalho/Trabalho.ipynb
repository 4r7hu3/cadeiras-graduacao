```python
import pandas as pd
```

```python
data = pd.read_excel('/home/arthur/Documentos/UFC/TPE/TPEdatabase.xlsx') # lê a planilha

# deleta algumas colunas
data.drop(['data/hora', 'concorda/discorda', 'cursoA', 
'cursoB', 'cursoC', 'cursoD',
'cursoE', 'cursoF', 'cursoG',
'cursoH', 'cursoI', 'cursoJ',
'cursoK', 'cursoL', 'cursoM',
'cursoN', 'cursoO', 'cursoP',
'cursoQ', 'cursoR', 'cursoS',
'visao_de_retorno_presencial'], axis='columns', inplace=True)

# deleta colunas/linhas totalmente vazias
data.dropna(how='all', inplace=True, axis=1)
data.dropna(how='all', inplace=True)
data.head() # lê as 5 primeiras linhas do dataframe
```

```python
# .value_counts() mostra a frenquência absoluta/relativa de cada resultado no dataframe; desconsidera valores nulos
# .describe() mostra uma descrição mais resumida, apresentando a moda, a quantidade de respostas não nulas (na/NaN) e quantos valores distintos há na coluna

testes = data['aspecto_testes_disponiveis'].value_counts(normalize=True)

rastreamento_de_casos = data['aspecto_rastreamento_de_casos'].value_counts(normalize=True)

espacos = data['aspecto_dimencionamento_de_espacos'].value_counts(normalize=True)

rus = data['aspecto_uso_dos_rus'].value_counts(normalize=True)

onibus_campus = data['aspecto_onibus_do_campus'].value_counts(normalize=True)

bilbiotecas = data['aspecto_funcionamento_das_bibliotecas'].value_counts(normalize=True)

passaporte_vacinal = data['posicionamento_exigencia_de_passaporte_vacinal'].describe() # como há 18 tipos diferentes de respostas, não é muito prático usar 'value_counts()'. Contudo, 'describe()' nos dá boas informações suficientes

deslocamento = data['deslocamento_entre_cidades'].value_counts(normalize=True)
```

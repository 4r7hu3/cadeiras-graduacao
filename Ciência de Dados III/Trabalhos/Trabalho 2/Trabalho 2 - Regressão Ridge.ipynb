## Implementação dos Algoritmos

```python
# biblioteca
import numpy as np
import random
```

### Funções de Qualidade do Ajuste

```python
# MSE (Erro Quadrático Médio)
def MSE(y_true, y_predict):
  n = len(y_true)
  SQRes = np.sum((y_true - y_predict)**2)

  mse = SQRes/n

  return mse
```

```python
# R2
def R2(y_true, y_predict):
  y_bar = np.mean(y_true)

  SQRes = np.sum((y_true - y_predict)**2)
  SQTot = np.sum((y_true - y_bar)**2)

  R_2 = 1 - SQRes/SQTot

  return R_2
```

```python
# MAPE (Erro Absoluto Percentual Médio)
def mape(y_true, y_predict):
  n = np.shape(y_true)[0]
  s = np.sum(np.abs((y_true - y_predict)/y_true))

  mape = s/n*100

  return mape
```

### Funções de Ajuste e Predição

```python
# regressão multivariada regularizada
def ridge_regression(X, y, lbd):

  n = np.shape(X)[0] # número de observações
  p = np.shape(X)[1] + 1 # número de variáveis + intercepto

  I = np.eye(p)
  I[0,0] = 0

  ones = np.ones(n)
  X = np.c_[ones, X]

  w = np.linalg.inv(np.transpose(X) @ X + lbd * I) @ np.transpose(X) @ y
  return w
```

```python
# função para predição
def predict_multivariate(X, w):
  n = np.shape(X)[0]
  ones = np.ones(n)

  X = np.c_[ones, X]

  return X @ w
```

## Experimento

### Carregando o conjunto de dados

```python
df = np.loadtxt('artificial1d.csv', delimiter=',')
print(df)
```

### Embaralhando o conjunto de dados

```python
np.random.seed(508492)
np.random.shuffle(df)
print(df)
```

### Separando as variáveis

```python
X = df[:, 0]
y = df[:, 1]

print(X)
print(y)
```

### Criando novas variáveis

```python
X2, X3, X4 = np.power(X, 2), np.power(X, 3), np.power(X, 4)

X = np.c_[X, X2, X3, X4]

print(X)
```

### Train/Test Split

```python
n = np.shape(X)[0]
i_final = round(n*0.7)

X_train = X[:i_final]
y_train = y[:i_final]
X_test  = X[i_final:]
y_test  = y[i_final:]

print(X_train)
print(y_train)
print(X_test)
print(y_test)
```

### Ajuste do modelo para $\lambda = 0$

```python
w = ridge_regression(X_train, y_train, 0)
print(w)
```

### Métricas do teste

```python
y_hat_test = predict_multivariate(X_test, w)

mse_test = MSE(y_test, y_hat_test)
r2_test = R2(y_test, y_hat_test)
mape_test = mape(y_test, y_hat_test)

print(mse_test)
print(r2_test)
print(mape_test)
```

### Ajuste com diferentes valores de $\lambda$

```python
ls = [0, 0.001, 0.01, 0.1, 1, 10, 100]

for i in range(len(ls)):
  w = ridge_regression(X_train, y_train, ls[i])

  # métricas de qualidade de ajuste p/ o conjunto de teste
  mse_t = MSE(y_test, predict_multivariate(X_test, w))
  r2_t = R2(y_test, predict_multivariate(X_test, w))
  mape_t = mape(y_test, predict_multivariate(X_test, w))

  print(f'\nAjuste com lambda = {ls[i]}\n')
  print(f'MSE: {mse_t}')
  print(f'R2: {r2_t}')
  print(f'MAPE: {mape_t}')
  print('\n--------------------------------')
```
